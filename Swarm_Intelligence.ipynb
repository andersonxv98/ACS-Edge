```python
import numpy as np
# import pyswarms as ps
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
from skimage import filters, feature
```

```python
img = (imread("https://drive.google.com/uc?id=1D_tAcJmO6AFfs66MgePS9xYiea2To8ZZ", as_gray=True) * 255).astype("uint8")
```

```python
imshow(img)
```

```python
# Aplicar filtro de Sobel
edges_sobel = filters.sobel(img)

# Aplicar detector de bordas Canny
edges_canny = feature.canny(img, sigma=1)
edges_canny_sig3 = feature.canny(img, sigma=3)

# Aplicar filtro de Prewitt
edges_prewitt = filters.prewitt(img)


plt.figure(figsize=(6, 6))
plt.imshow(edges_sobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(edges_canny, cmap='gray')
plt.title('Canny (sigma=1)')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(edges_canny_sig3, cmap='gray')
plt.title('Canny (sigma=3)')
plt.axis('off')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(edges_prewitt, cmap='gray')
plt.title('Prewitt')
plt.axis('off')
plt.show()
```
